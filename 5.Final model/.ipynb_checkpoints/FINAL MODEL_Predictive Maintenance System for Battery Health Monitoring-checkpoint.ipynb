# Predictive Maintenance System for Battery Health Monitoring
1.Example: Serial number or entry number.

2.Battery Type: Type or model of the battery.

3.Battery Voltage Range: Range of voltage the battery operates within.

4.Load Testing: Result of load testing (Pass/Fail/Categorical).

5.Visual Inspection: Result of visual inspection (Yes/No/Categorical).

6.Age of Battery: Age of the battery in years and months.

7.Maintenance History: History of maintenance activities performed.

8.Specific Gravity: Specific gravity of the battery electrolyte.

9.OCV (Open Circuit Voltage): Voltage when battery terminals are disconnected.

10.CCA (Cold Cranking Amps): Measure of the battery's ability to start an engine in cold temperatures.

11.Status: Overall status of the battery (OK/Rejected/Categorical).

12.Maintenance Required: Indicates whether maintenance is required for the battery (Required/Not Required/Categorical).

```python
dataset = pd.read_csv("preprocessed_Battery_data.csv", index_col=None)
df2 = dataset
df2 = pd.get_dummies(df2, drop_first=True)

indep_X = df2.drop(['Maintenance Required_Yes', 'Charge Needed_Yes', 'Status_Low Charge', 'Status_Weak Battery'], axis=1)
dep_Y = df2['Maintenance Required_Yes']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

class ModelEvaluator:
    def __init__(self):
        pass
    
    def split_scalar(self, indep_X, dep_Y):
        X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.25, random_state=0)

        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)    
        return X_train, X_test, y_train, y_test

    def accuracy(self, classifier, X_train, y_train, X_test, y_test):
        classifier.fit(X_train, y_train)
        y_pred = classifier.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        return acc

    def decision_tree(self, X_train, y_train, X_test, y_test):
        from sklearn.tree import DecisionTreeClassifier
        classifier = DecisionTreeClassifier(random_state=0)
        return self.accuracy(classifier, X_train, y_train, X_test, y_test)

    def random_forest(self, X_train, y_train, X_test, y_test):
        from sklearn.ensemble import RandomForestClassifier
        classifier = RandomForestClassifier(n_estimators=10, random_state=0)
        return self.accuracy(classifier, X_train, y_train, X_test, y_test)

    def xgboost(self, X_train, y_train, X_test, y_test):
        from xgboost import XGBClassifier
        classifier = XGBClassifier(n_jobs=5, learning_rate=0.1, max_depth=10, random_state=1)
        return self.accuracy(classifier, X_train, y_train, X_test, y_test)

# Example usage:
evaluator = ModelEvaluator()
X_train, X_test, y_train, y_test = evaluator.split_scalar(indep_X, dep_Y)

# Decision Tree
dt_accuracy = evaluator.decision_tree(X_train, y_train, X_test, y_test)
print("Decision Tree Accuracy:", dt_accuracy)

# Random Forest
rf_accuracy = evaluator.random_forest(X_train, y_train, X_test, y_test)
print("Random Forest Accuracy:", rf_accuracy)

# XGBoost
xgb_accuracy = evaluator.xgboost(X_train, y_train, X_test, y_test)
print("XGBoost Accuracy:", xgb_accuracy)
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def split_scalar(indep_X, dep_Y):
    X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.25, random_state=0)
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)    
    return X_train, X_test, y_train, y_test

def accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred)

def classification_accuracy(classifier, X_train, y_train, X_test, y_test):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    return accuracy(y_test, y_pred)

def rfeFeature(indep_X, dep_Y, n):
    classifiers = {
        "Logistic Regression": LogisticRegression(),
        "Decision Tree": DecisionTreeClassifier(random_state=0),
        "Random Forest": RandomForestClassifier(n_estimators=10, random_state=0),
        "XGBoost": XGBClassifier(n_jobs=5, learning_rate=0.1, max_depth=10, random_state=1)
    }

    rfelist = []
    colnames_list = []  
    accuracies = []  

    for model_name, classifier in classifiers.items():
        log_rfe = RFE(estimator=classifier, n_features_to_select=n)
        log_fit = log_rfe.fit(indep_X, dep_Y)
        log_rfe_feature = log_fit.transform(indep_X)
        rfelist.append(log_rfe_feature)

        selected_columns = [col for col, selected in zip(indep_X.columns, log_rfe.support_) if selected]
        colnames_list.append(selected_columns)

        X_train, X_test, y_train, y_test = split_scalar(pd.DataFrame(log_rfe_feature), dep_Y)
        acc = classification_accuracy(classifier, X_train, y_train, X_test, y_test)
        accuracies.append(acc)

    return rfelist, colnames_list, accuracies

# Call the function with your data
rfelist, colnames_list, accuracies = rfeFeature(indep_X, dep_Y, 5)

# Print the selected column names and accuracies for each model
for model_name, selected_columns, accuracy in zip(["Logistic Regression", "Decision Tree", "Random Forest", "XGBoost"], colnames_list, accuracies):
    print(f"Model: {model_name}")
    print("Selected Columns:", selected_columns)
    print(f"Accuracy: {accuracy}\n")
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def split_scalar(indep_X, dep_Y):
    X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.25, random_state=0)
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)    
    return X_train, X_test, y_train, y_test

def calculate_accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred)

def classification_accuracy(classifier, X_train, y_train, X_test, y_test):
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    return calculate_accuracy(y_test, y_pred)

def rfeFeature(indep_X, dep_Y, n):
    classifiers = {
        "Logistic Regression": LogisticRegression(),
        "Decision Tree": DecisionTreeClassifier(random_state=0),
        "Random Forest": RandomForestClassifier(n_estimators=10, random_state=0),
        "XGBoost": XGBClassifier(n_jobs=5, learning_rate=0.1, max_depth=10, random_state=1)
    }

    rfelist = []
    colnames_list = []  
    accuracies = []  

    for model_name, classifier in classifiers.items():
        log_rfe = RFE(estimator=classifier, n_features_to_select=n)
        log_fit = log_rfe.fit(indep_X, dep_Y)
        log_rfe_feature = log_fit.transform(indep_X)
        rfelist.append(log_rfe_feature)

        selected_columns = [col for col, selected in zip(indep_X.columns, log_rfe.support_) if selected]
        colnames_list.append(selected_columns)

        X_train, X_test, y_train, y_test = split_scalar(pd.DataFrame(log_rfe_feature), dep_Y)
        acc = classification_accuracy(classifier, X_train, y_train, X_test, y_test)
        accuracies.append(acc)

    return rfelist, colnames_list, accuracies

# Call the function with your data
rfelist, colnames_list, accuracies = rfeFeature(indep_X, dep_Y, 5)

# Print the selected column names and accuracies for each model
for model_name, selected_columns, accuracy in zip(["Logistic Regression", "Decision Tree", "Random Forest", "XGBoost"], colnames_list, accuracies):
    print(f"Model: {model_name}")
    print("Selected Columns:", selected_columns)
    print(f"Accuracy: {accuracy}\n")

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

# Call the function with your data to obtain rfelist
rfelist= rfeFeature(indep_X, dep_Y, 5)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.30, random_state=0)

# Transform the data using the selected features obtained from RFE
# Define a list to store transformed feature arrays
transformed_features = []

# Iterate over each transformed feature array in rfelist
for rfefeature in rfelist:
    # Transform the features using the selected columns
    transformed_feature = rfefeature.transform(X_train)
    transformed_features.append(transformed_feature)

# Concatenate the transformed features
X_train_rfe = np.concatenate(transformed_features, axis=1)

# Repeat the same process for the test set
transformed_features_test = []
for rfefeature_test in rfelist:
    transformed_feature_test = rfefeature_test.transform(X_test)
    transformed_features_test.append(transformed_feature_test)
X_test_rfe = np.concatenate(transformed_features_test, axis=1)

# Define the parameter grid for GridSearchCV
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'criterion': ['gini', 'entropy'],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': ['balanced', 'balanced_subsample']
}

# Create the RandomForestClassifier instance
rf_classifier = RandomForestClassifier()

# Create GridSearchCV object for Random Forest
grid_search_rf = GridSearchCV(rf_classifier, param_grid_rf, cv=5, scoring='accuracy')

# Fit the grid search to find the best parameters
grid_search_rf.fit(X_train_rfe, y_train)

# Print the best parameters found by GridSearchCV
print("Best Parameters for Random Forest:", grid_search_rf.best_params_)

# Get the best classifier with the best parameters
best_classifier_rf = grid_search_rf.best_estimator_

# Train the best classifier on the entire training data
best_classifier_rf.fit(X_train_rfe, y_train)

# Predictions on the test set
y_pred_rf = best_classifier_rf.predict(X_test_rfe)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Accuracy for Random Forest:", accuracy_rf)

# Other evaluation metrics
print("Classification Report for Random Forest:")
print(classification_report(y_test, y_pred_rf))

print("Confusion Matrix for Random Forest:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv("preprocessed_Battery_data.csv", index_col=None)
dataset= pd.get_dummies(dataset, drop_first=True)
X= dataset.drop(['Maintenance Required_Yes', 'Charge Needed_Yes', 'Status_Low Charge', 'Status_Weak Battery'], axis=1)
y= dataset['Maintenance Required_Yes']


# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Applying RFE
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
rfe =RFE(estimator=rf, n_features_to_select=5)
X_train = rfe.fit_transform(X_train, y_train)
X_test = rfe.transform(X_test)

# Fitting Logistic Regression to the Training set
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators' :[100], 'criterion':['gini', 'entropy', 'log_loss'],
           'max_features':['sqrt', 'log2'], 'class_weight':['balanced', 'balanced_subsample']}


grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
re=grid.cv_results_
y_predict=grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_predict)
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,y_predict,average='weighted')
print("The f1 macro value the best parameter{}:".format(grid.best_params_),f1_macro)
```
