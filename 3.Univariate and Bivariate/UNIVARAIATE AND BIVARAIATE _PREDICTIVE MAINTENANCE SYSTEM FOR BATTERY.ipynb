# Predictive Maintenance System for Battery Health Monitoring
1.Serial Number: Sequential number assigned to each battery.

2.Battery Voltage (V): The voltage of the battery measured in volts (V).

3.Visual Inspection: The result of the visual inspection performed on the battery, indicating any visible damage or irregularities.

4.Starting: The condition of the battery's performance during startup, such as smooth cranking or struggling cranking.

5.Lights and Electronics: The behavior of lights and electronic devices powered by the battery, indicating normal functioning or issues like flickering lights.

7.Battery Age (Years): The age of the battery expressed in years.

8.Battery Age (Months): The age of the battery expressed in months.

10.Maintenance History: Details about the maintenance performed on the battery, including regular maintenance or neglected maintenance.

11.Specific Gravity: The specific gravity of the battery's electrolyte.

12.Capacity Range (Ah): The range of capacity of the battery measured in ampere-hours (Ah).

13.Open Circuit Voltage (V): The voltage of the battery measured when it is not connected to any load.

14.Acid Level Check: The result of checking the acid level of the battery, indicating normal or low levels.

15.Charge Needed: Whether the battery requires charging or not.

16.Maintenance Required: Indicates if maintenance is required based on the battery's condition.

17.Status: Overall status of the battery, such as "Battery OK" or "Weak Battery".

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99","Q4:100%","IQR",
                               "1.5Rule","Lesser","Greater","Min","Max"],columns=quan)
for coulmnName in quan:
    descriptive[coulmnName]["Mean"]=dataset[coulmnName].mean()
    descriptive[coulmnName]["Median"]=dataset[coulmnName].median()
    descriptive[coulmnName]["Mode"]=dataset[coulmnName].mode()[0]
    descriptive[coulmnName]["Q1:25%"]=dataset.describe()[coulmnName]["25%"]
    descriptive[coulmnName]["Q2:50%"]=dataset.describe()[coulmnName]["50%"]
    descriptive[coulmnName]["Q3:75%"]=dataset.describe()[coulmnName]["75%"]
    descriptive[coulmnName]["99"]=np.percentile(dataset[coulmnName],99)
    descriptive[coulmnName]["Q4:100%"]=dataset.describe()[coulmnName]["max"]
    descriptive[coulmnName]["IQR"]= descriptive[coulmnName]["Q3:75%"]-descriptive[coulmnName]["Q1:25%"]
    descriptive[coulmnName]["1.5Rule"]=1.5*descriptive[coulmnName]["IQR"]
    descriptive[coulmnName]["Lesser"]=descriptive[coulmnName]["Q1:25%"]-descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Greater"]=descriptive[coulmnName]["Q3:75%"]+ descriptive[coulmnName]["1.5Rule"]
    descriptive[coulmnName]["Min"]=dataset[coulmnName].min()
    descriptive[coulmnName]["Max"]=dataset[coulmnName].max()
```

```python
def Univariate(dataset,quan):
    descriptive=pd.DataFrame(index=["MEAN","MEDIAN","MODE","Q1:25%","Q2:50%","Q3:75%",
                                    "99%","Q4:100%","IQR","1.5 rule",
                                    "Lesser","Greater","Min","Max","skew","kurtosis","var","std"],columns=quan)
    for columName in quan:
        descriptive[columName]["MEAN"]=dataset[columName].mean()
        descriptive[columName]["MEDIAN"]=dataset[columName].median()
        descriptive[columName]["MODE"]=dataset[columName].mode()[0]
        descriptive[columName]["Q1:25%"]=dataset.describe()[columName]["25%"]
        descriptive[columName]["Q2:50%"]=dataset.describe()[columName]["50%"]
        descriptive[columName]["Q3:75%"]=dataset.describe()[columName]["75%"]
        descriptive[columName]["99%"]=np.percentile(dataset[columName],99)
        descriptive[columName]["Q4:100%"]=dataset.describe()[columName]["max"]
        descriptive[columName]["IQR"]=descriptive[columName]["Q3:75%"]- descriptive[columName]["Q1:25%"]
        descriptive[columName]["1.5 rule"]=1.5* descriptive[columName]["IQR"]
        descriptive[columName]["Lesser"]=descriptive[columName]["Q1:25%"]- descriptive[columName]["1.5 rule"]
        descriptive[columName]["Greater"]=descriptive[columName]["Q3:75%"]+ descriptive[columName]["1.5 rule"]
        descriptive[columName]["Min"]=dataset[columName].min()
        descriptive[columName]["Max"]=dataset[columName].max()
        descriptive[columName]["skew"]=dataset[columName].skew()
        descriptive[columName]["kurtosis"]=dataset[columName].kurtosis()
        descriptive[columName]["var"]=dataset[columName].var()
        descriptive[columName]["std"]=dataset[columName].std()
    return descriptive
    
```

# Bivariate Analysis

# 1.Is there a relationship between battery voltage and visual inspection results?

```python

# Box plot of voltage distribution for different visual inspection results
plt.figure(figsize=(8, 6))
sns.boxplot(x='Visual Inspection', y='Battery Voltage (V)', data=dataset)
plt.title('Voltage Distribution by Visual Inspection Result')
plt.xlabel('Visual Inspection Result')
plt.ylabel('Battery Voltage (V)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# 2.Is there a correlation between battery age and the need for charging?

```python
# Convert 'Charge Needed' column to numerical values
dataset['Charge Needed'] = dataset['Charge Needed'].map({'Yes': 1, 'No': 0})

# Now compute the correlation coefficient
correlation = dataset['Battery Age (Years)'].corr(dataset['Charge Needed'])

print("Pearson Correlation Coefficient:", correlation)
```

# 3.How does the maintenance status vary concerning the battery's starting condition?

```python
# Assuming 'dataset' contains your data
# Let's rename 'Starting' column to avoid spaces
dataset.rename(columns={'Starting': 'Starting Condition'}, inplace=True)

# Create a contingency table
contingency_table = pd.crosstab(dataset['Maintenance History'], dataset['Starting Condition'])

# Plotting the contingency table as a grouped bar plot
contingency_table.plot(kind='bar', figsize=(10, 6))
plt.title('Maintenance Status vs Starting Condition')
plt.xlabel('Maintenance Status')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Starting Condition')
plt.show()
```

# 4.Does battery voltage correlate with the requirement for maintenance?

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dataset['Battery Voltage (V)'], dataset['Maintenance Required'], alpha=0.5)
plt.title('Battery Voltage vs Maintenance Requirement')
plt.xlabel('Battery Voltage (V)')
plt.ylabel('Maintenance Required')
plt.grid(True)
plt.show()
```

# 5.Is there an association between battery age and the need for professional evaluation?

```python
# Create a contingency table
contingency_table = pd.crosstab(dataset['Battery Age (Years)'], dataset['Maintenance Required'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the chi-square statistic and p-value
print("Chi-square Statistic:", chi2)
print("p-value:", p)
```

# 6.How does the distribution of battery status vary concerning maintenance history?

```python
# Create a contingency table
contingency_table = pd.crosstab(dataset['Maintenance History'], dataset['Status'])

# Plot stacked bar plot
contingency_table.plot(kind='bar', stacked=True)
plt.xlabel('Maintenance History')
plt.ylabel('Count')
plt.title('Distribution of Battery Status by Maintenance History')
plt.legend(title='Battery Status')
plt.xticks(rotation=45)
plt.show()
```
